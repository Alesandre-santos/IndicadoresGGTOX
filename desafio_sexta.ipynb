```python
import requests
import pandas as pd
import io

URL = "https://dados.anvisa.gov.br/dados/VigiMed_Reacoes.csv"

urlData = requests.get(URL, verify=False).content

dfrea = pd.read_csv(io.StringIO(urlData.decode('iso-8859-1')),sep=';', error_bad_lines=False, decimal=",")
display(dfrea)
```
